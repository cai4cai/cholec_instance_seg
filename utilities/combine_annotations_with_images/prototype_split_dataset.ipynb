## Combine the datasets

```python
import os
import shutil
```

```python
def count_files_in_directory(directory):
    file_count = 0
    for root, _, files in os.walk(directory):
        file_count += len(files)
    return file_count
```

```python
# Define the directories
reference_image_dataset  = "../../../datasets/cholecinstance_seg_reference_image_set"
# Define target directory for annotations, change  to where it is in your repository 
annotation_dataset  = "../../../datasets/cholecinstanceseg"
```

```python
# Iterate through dataset splits (train, validation, etc.)
for split in os.listdir(annotation_dataset):
    split_ann_path = os.path.join(annotation_dataset, split)
    split_orig_path = os.path.join(reference_image_dataset, split)

    if os.path.isdir(split_ann_path):  # Ensure it's a directory
        # Iterate through video folders in the annotation dataset
        for video_folder in os.listdir(split_ann_path):
            ann_source_path = os.path.join(split_ann_path, video_folder, "ann_dir")
            video_orig_path = os.path.join(split_orig_path, video_folder)
            ann_target_path = os.path.join(video_orig_path, "ann_dir")

            if os.path.exists(ann_source_path):  # Ensure ann_dir exists
                os.makedirs(video_orig_path, exist_ok=True)  # Ensure video folder exists
                shutil.move(ann_source_path, ann_target_path)
                print(f"Moved: {ann_source_path} -> {ann_target_path}")
            else:
                print(f"Skipping {ann_source_path}, does not exist.")


# Check if annotation dataset is empty and delete it
if count_files_in_directory(annotation_dataset) == 0:
    shutil.rmtree(annotation_dataset)
    print(f"Deleted empty directory: {annotation_dataset}")
    
#rename dataset. 

# Get the absolute path of the original dataset
parent_dir = os.path.abspath(os.path.join(reference_image_dataset, os.pardir))

# Construct the new full path
new_dataset_path = os.path.join(parent_dir, 'cholecinstanceseg_reference_image_and_annotations')

# Rename the directory
if os.path.exists(reference_image_dataset):  # Ensure the dataset exists
    os.rename(reference_image_dataset, new_dataset_path)
    print(f"Renamed: {reference_image_dataset} -> {new_dataset_path}")
else:
    print(f"Error: {reference_image_dataset} does not exist.")
 

print("All annotation directories have been restored successfully!")
```
